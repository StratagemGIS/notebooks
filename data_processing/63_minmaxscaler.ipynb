```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.DataFrame([
    [-1000, 200],
    [100, -800],
    [700, 100],
    [400, 1800]
], columns=['c1', 'c2'])
```

```python
scaler = MinMaxScaler(feature_range=(0, 1)).set_output(transform='pandas')
scaled = scaler.fit_transform(data)
scaled
```

```python
print(scaler.data_min_, scaler.data_max_)
```
