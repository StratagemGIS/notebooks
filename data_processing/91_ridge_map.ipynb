```python
import matplotlib.pyplot as plt
from ridge_map import RidgeMap
```

```python
rm = RidgeMap((-66.986, 47.286, -66.667, 47.527))

values = rm.get_elevation_data(num_lines=300)

values=rm.preprocess(
    values=values,
    lake_flatness=2,
    water_ntile=10,
    vertical_ratio=240
)

rm.plot_map(
    values=values,
    label='Mount Carleton \nProvincial Park',
    label_y=0.1,
    label_x=0.55,
    label_size=40,
    linewidth=2
)

plt.savefig('ridge_map.png', bbox_inches='tight', dpi=600);
```
