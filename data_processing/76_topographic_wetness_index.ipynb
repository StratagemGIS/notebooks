```python
import numpy as np
import pooch
import richdem as rd
```

```python
dem_file = pooch.retrieve(
    'https://geonb.snb.ca/downloads2/elevation/20162017/dem/geotiff/nb_2016_2017_dem_2470000_7590000.tif',
    known_hash='5239d1387cbc3a3e62e89ad7afd9ad0a9443cf4cb28aeafc87ee43ced6eec4eb'
)
```

```python
dem = rd.LoadGDAL(dem_file)
fa = rd.FlowAccumulation(dem, method='D8')
slope = rd.TerrainAttribute(dem, attrib='slope_radians')
twi = np.log(fa / np.tan(slope))
```

```python
rd.rdShow(twi, axes=False, cmap='jet', figsize=(8, 5.5));
```
