```python
import pandas as pd
from faker import Faker
```

```python
Faker.seed(2)
fake = Faker()

df = pd.DataFrame({
    'name': [fake.name() for i in range(5)],
    'email': [fake.email() for i in range(5)],
    'country': [fake.country() for i in range(5)],
    'username': [fake.user_name() for i in range(5)],
    'dob': [fake.date_of_birth() for i in range(5)]
})

df
```
