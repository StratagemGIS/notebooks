```python
import matplotlib.pyplot as plt
import pooch
import rasterio as rio
import richdem as rd
```

```python
dem_file = pooch.retrieve(
    'https://geonb.snb.ca/downloads2/elevation/20162017/dem/geotiff/nb_2016_2017_dem_2470000_7590000.tif',
    known_hash='5239d1387cbc3a3e62e89ad7afd9ad0a9443cf4cb28aeafc87ee43ced6eec4eb'
)
```

```python
with rio.open(dem_file) as src:
    dem = src.read(1)
    rd_dem = rd.rdarray(dem, no_data=src.nodata)
    slope = rd.TerrainAttribute(rd_dem, attrib='slope_riserun')
    aspect = rd.TerrainAttribute(rd_dem, attrib='aspect')
```

```python
rd.rdShow(rd_dem, axes=False, cmap='jet', figsize=(8, 5.5));
```

```python
rd.rdShow(slope, axes=False, cmap='jet', figsize=(8, 5.5));
```

```python
rd.rdShow(aspect, axes=False, cmap='jet', figsize=(8, 5.5));
```
