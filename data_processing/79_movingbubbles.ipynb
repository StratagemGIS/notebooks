```python
import numpy as np
import pandas as pd
from d3blocks import D3Blocks
```

```python
# @title Dataset Preparation

np.random.seed(2024)

n = 250  # Number of Students

activities = [
    'Attending Classes',
    'Studying',
    'Socializing',
    'Dining',
    'Exercising',
    'Volunteering',
    'Relaxing',
    'Group Projects',
    'Attending Events'
]

datetime = (
    pd.to_datetime('2025-01-01') +
    pd.to_timedelta(np.random.rand(n * len(activities)) * 86400, unit='s')
)

students_df = pd.DataFrame(
    {
        'datetime': datetime,
        'sample_id': np.random.choice(range(0, n), size=n * len(activities)),
        'state': np.random.choice(activities, size=n * len(activities))
     }
)
```

```python
students_df.sample(5, random_state=2024)
```

```python
d3 = D3Blocks(chart='movingbubbles')
d3.movingbubbles(students_df, figsize=(775, 800))
```
