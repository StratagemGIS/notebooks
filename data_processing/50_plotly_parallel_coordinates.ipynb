```python
import pandas as pd
import plotly.express as px
import pooch
```

```python
spotify_songs = pooch.retrieve(
    'https://github.com/StratagemGIS/datasets/raw/main/table/spotify_songs.csv',
    known_hash='9902920cfe4940f0914ebb6716801a42f7f19ac88fbd6693eaae7e2822b16845'
)
```

```python
df = pd.read_csv(spotify_songs, engine='pyarrow', dtype_backend='pyarrow')
```

```python
fig = px.parallel_coordinates(
    pd.concat([df.select_dtypes(float), df.track_popularity], axis=1),
    color='track_popularity',
    color_continuous_scale='Spectral',
)

fig.show()
```
