```python
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
```

```python
import geopandas as gpd
import pooch
from tesspy import Tessellation
```

```python
aoi_file = pooch.retrieve(
    'https://github.com/StratagemGIS/datasets/raw/main/vector/City_Boundary.zip',
    known_hash='431710489ce01850be3ed0ac816305b41cbd871b14fae7d3c94bb8da469c6aa6'
)
aoi = gpd.read_file(aoi_file)
fredericton = Tessellation(aoi.assign(osm_id=''))
```

```python
m = aoi.explore(width=500, height=500, color='red')
fredericton.squares(resolution=14).explore(m=m)
```

```python
m = aoi.explore(width=500, height=500, color='red')
fredericton.hexagons(resolution=8).explore(m=m)
```

```python
m = aoi.explore(width=500, height=500, color='red')
fredericton.voronoi().explore(m=m)
```
