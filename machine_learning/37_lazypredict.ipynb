```python
import pandas as pd
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
penguins = pd.read_csv(
    'https://raw.githubusercontent.com/StratagemGIS/datasets/main/table/penguins.csv',
)
```

```python
x = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].values
y = penguins.species.map({'Adelie': 0, 'Gentoo': 1, 'Chinstrap': 2}).values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20240401)
clf = LazyClassifier(verbose=0)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
```

```python
models.drop(columns=['ROC AUC'])
```
